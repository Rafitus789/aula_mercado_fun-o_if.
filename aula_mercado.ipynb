```python
#o desafio é mostrar os produtos de um mercado
#quero poder escolher os produtos, para conseguir informações sobre eles
#quero poder adicionar novos produtos
#quero poder excluir um produto do mercado

#-----------

#criar as variaveis globais para o software
lista_produtos = ['Detergente Ype',
                 'Refrigerante Gengibirra',
                 'Arroz',
                 'Feijão',
                 'Água Mineral']

lista_preco = [2.50, 5, 10, 9, 2]

list_products = []
list_price = []
list_description = []
```

```python
#para criarmos uma função devemos armazenas os códigos dentro de um bloco e dar um nome
#funcao 1 - ver produtos
def ver_produtos():
  print(f'Os produtos que temos são: {lista_produtos}')
  print(f'e seus preços são: {lista_preco}')

def adicionar_produtos():
  novo_produto = input('Informe o nome do produto: ')
  novo_preco = float(input('Informe o preço: '))

  # Precisamos adicionar esses novos valores a lista inicial de produtos e preços
  lista_produtos.append(novo_produto)
  lista_preco.append(novo_preco)
```

```python
print('---- MENU DO MERCADO-----')
print('Escolha a opção:')
print('1 - ver produtos')
print('2 - cadastrar novo produto')
opcao = int(input('informe a opção desejada: '))

if opcao == 1:
  ver_produtos()
elif opcao ==2:
  adicionar_produtos()
  ver_produtos()
else:
  print('Digite uma opção válida')
```
